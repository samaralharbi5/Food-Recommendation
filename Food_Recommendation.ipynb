```python
import pandas as pd
import random

# Meal types
meal_types = ['Breakfast', 'Lunch', 'Dinner', 'Snacks']
meals = ['Oatmeal with fruits', 'Green salad', 'Grilled chicken with vegetables', 'Yogurt with almonds',
         'Grilled fish', 'Egg toast', 'Light appetizers', 'Hummus salad', 'Tuna salad', 'Lentil soup']

# Create a mock database with 200 meals
data = []

for meal_type in meal_types:
    for i in range(50):
        meal = random.choice(meals)  # Choose a random meal from the list
        sodium_level = random.randint(50, 300)  # Sodium level (mg/100g)
        glucose_level = random.randint(10, 100)  # Glucose level (mg/100g)
        data.append([meal, meal_type, sodium_level, glucose_level])

# Create DataFrame
df = pd.DataFrame(data, columns=['Meal', 'Meal Type', 'Sodium Level', 'Glucose Level'])

# Save the database to a CSV file
df.to_csv('meal_recommendations.csv', index=False)

print("Mock database created successfully!")
```

```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# تحميل قاعدة البيانات
df = pd.read_csv('meal_recommendations.csv')

# دالة لاختيار الوجبات بناءً على المدخلات
def get_recommendations():
    try:
        glucose_level = int(entry_glucose.get()) if entry_glucose.get() else 0
        sodium_level = int(entry_sodium.get()) if entry_sodium.get() else 0
        
        # تصفية الوجبات بناءً على المدخلات
        if glucose_level and sodium_level:
            recommendations = df[(df['Glucose Level'] <= glucose_level) & (df['Sodium Level'] <= sodium_level)]
        elif glucose_level:
            recommendations = df[df['Glucose Level'] <= glucose_level]
        elif sodium_level:
            recommendations = df[df['Sodium Level'] <= sodium_level]
        else:
            recommendations = df  # إذا لم يتم إدخال أي شيء، عرض كل الوجبات
        
        if recommendations.empty:
            messagebox.showinfo("No Recommendations", "No meals match your criteria.")
        else:
            output_text.delete(1.0, tk.END)  # مسح النص الحالي
            for _, row in recommendations.iterrows():
                output_text.insert(tk.END, f"Meal: {row['Meal']}, Type: {row['Meal Type']}\n")
    
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numeric values for glucose and sodium.")

# دالة لتحديد الحقول التي ستظهر بناءً على الإجابة
def show_inputs():
    condition = var_condition.get()
    
    # إخفاء الحقول السابقة
    for widget in frame_inputs.winfo_children():
        widget.grid_forget()
    
    # إظهار المدخلات المناسبة بناءً على الحالة
    if condition == "Diabetes":
        label_glucose.grid(row=0, column=0, padx=10, pady=10)
        entry_glucose.grid(row=0, column=1, padx=10, pady=10)
    elif condition == "Hypertension":
        label_sodium.grid(row=0, column=0, padx=10, pady=10)
        entry_sodium.grid(row=0, column=1, padx=10, pady=10)
    elif condition == "Both":
        label_glucose.grid(row=0, column=0, padx=10, pady=10)
        entry_glucose.grid(row=0, column=1, padx=10, pady=10)
        label_sodium.grid(row=1, column=0, padx=10, pady=10)
        entry_sodium.grid(row=1, column=1, padx=10, pady=10)
    else:
        messagebox.showerror("Error", "Please select a health condition.")

# إنشاء واجهة المستخدم
root = tk.Tk()
root.title("Food Recommendation System for Diabetic and Hypertension Patients")
root.geometry("400x400")  # تغيير الحجم

# إضافة تنسيق للواجهة
root.configure(bg="#f0f8ff")

# سؤال عن الحالة الصحية
frame_condition = tk.Frame(root, bg="#f0f8ff")
frame_condition.pack(pady=20)

label_condition = tk.Label(frame_condition, text="Do you have Diabetes, Hypertension, or Both?", bg="#f0f8ff", font=("Arial", 12))
label_condition.pack()

var_condition = tk.StringVar()
var_condition.set("Select")

option_menu = tk.OptionMenu(frame_condition, var_condition, "Diabetes", "Hypertension", "Both")
option_menu.config(width=20)
option_menu.pack(pady=10)

button_condition = tk.Button(frame_condition, text="Next", command=show_inputs, bg="#4CAF50", fg="white", font=("Arial", 12))
button_condition.pack()

# Frame لإدخال البيانات (جلوكوز وصوديوم)
frame_inputs = tk.Frame(root, bg="#f0f8ff")
frame_inputs.pack(pady=20)

label_glucose = tk.Label(frame_inputs, text="Enter Glucose Level (mg):", bg="#f0f8ff", font=("Arial", 12))
entry_glucose = tk.Entry(frame_inputs, font=("Arial", 12))

label_sodium = tk.Label(frame_inputs, text="Enter Sodium Level (mg):", bg="#f0f8ff", font=("Arial", 12))
entry_sodium = tk.Entry(frame_inputs, font=("Arial", 12))

# زر للحصول على التوصيات
button_get_recommendations = tk.Button(root, text="Get Recommendations", command=get_recommendations, bg="#008CBA", fg="white", font=("Arial", 12))
button_get_recommendations.pack(pady=10)

# مساحة لعرض التوصيات
output_text = tk.Text(root, height=10, width=50, font=("Arial", 12))
output_text.pack()

# تشغيل الواجهة
root.mainloop()
```
